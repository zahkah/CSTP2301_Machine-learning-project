```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso, Ridge, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Paths for datasets and test sets
dataset_paths = ['./content/534541-corrected.csv', './content/534546-corrected.csv', './content/549305-corrected.csv']
test_paths = ['./test set/534541_test.csv', './test set/534546_test.csv', './test set/549305_test.csv']
target_columns = ['534541', '534546', '549305']

# Models and their configurations
models = {
    "Lasso Regression": {"model": Lasso(max_iter=10000, random_state=42), "params": {"alpha": [0.001, 0.01, 0.1, 1, 10]}},
    "Linear Regression": {"model": LinearRegression(), "params": None},
    "Random Forest": {
        "model": RandomForestRegressor(random_state=42),
        "params": {
            "n_estimators": [100, 200],
            "max_depth": [10, 20, None],
            "min_samples_split": [2, 5],
            "min_samples_leaf": [1, 2],
        },
    },
    "Ridge Regression": {"model": Ridge(random_state=42), "params": {"alpha": [0.01, 0.1, 1, 10]}},
    "Decision Tree": {
        "model": DecisionTreeRegressor(random_state=42),
        "params": {
            "max_depth": [5, 10, 20, None],
            "min_samples_split": [2, 5],
            "min_samples_leaf": [1, 2],
        },
    },
    "Support Vector Regressor": {
        "model": SVR(),
        "params": {
            "C": [0.1, 1, 10, 100],
            "epsilon": [0.01, 0.1, 1],
            "gamma": ["scale", "auto"],
            "kernel": ["linear", "rbf"],
        },
    },
    "MLP Regressor": {
        "model": MLPRegressor(random_state=42, max_iter=2000, early_stopping=True, validation_fraction=0.2),
        "params": {
            "hidden_layer_sizes": [(50,), (100,), (50, 50)],
            "alpha": [0.0001, 0.001, 0.01],
            "learning_rate": ["constant", "adaptive"],
            "batch_size": [16, 32],
        },
    },
    "XGBoost": {
        "model": xgb.XGBRegressor(objective="reg:squarederror", random_state=42),
        "params": {
            "learning_rate": [0.01, 0.05, 0.1],
            "max_depth": [3, 5, 7],
            "n_estimators": [50, 100, 200],
            "subsample": [0.8, 1.0],
            "colsample_bytree": [0.8, 1.0],
            "reg_alpha": [0, 0.1],
            "reg_lambda": [1, 10],
        },
    },
}

# Summary results
summary_results = []

# Loop through datasets and models
best_models = {}




for dataset_path, target_column, test_path in zip(dataset_paths, target_columns, test_paths):
    # Load dataset
    data = pd.read_csv(dataset_path)
    data.fillna(data.mean(), inplace=True)


        # Extract the 'date' column and drop it from the features
    dates = data['date']
    X = data.drop(columns=[target_column, 'date'])  # Remove 'date' before scaling and transforming
    y = data[target_column]

    # Log-transform target if skewed
    log_transform = False
    if np.abs(y.skew()) > 0.5:
        log_transform = True
        y = np.log1p(y)

    # Normalize and add polynomial features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    poly = PolynomialFeatures(degree=2, include_bias=False)
    X_poly = poly.fit_transform(X_scaled)

    # Split dataset into training, validation, and test sets
    X_train, X_val, X_test = X_poly[:-48], X_poly[-48:-12], X_poly[-12:]
    y_train, y_val, y_test = y[:-48], y[-48:-12], y[-12:]
    train_dates, val_dates, test_dates = dates[:-48], dates[-48:-12], dates[-12:]

    # Run each model and record results
    dataset_results = {"Target ID": target_column}
    print(f"\nProcessing dataset: {target_column}")
    for model_name, config in models.items():
        model = config["model"]
        params = config["params"]

        # Grid/Randomized Search for hyperparameter tuning
        if params:
            search = GridSearchCV if model_name in ["Lasso Regression", "Ridge Regression", "Decision Tree"] else RandomizedSearchCV
            if search == GridSearchCV:
                grid_search = GridSearchCV(model, params, scoring="neg_mean_squared_error", cv=3, n_jobs=-1)
            else:
                grid_search = RandomizedSearchCV(model, params, scoring="neg_mean_squared_error", cv=3, n_jobs=-1, random_state=42)
            grid_search.fit(X_train, y_train)
            model = grid_search.best_estimator_

        # Train model on training set
        model.fit(X_train, y_train)
        y_val_pred = model.predict(X_val)

        # Validation metrics
        val_mse = mean_squared_error(y_val, y_val_pred)
        val_mae = mean_absolute_error(y_val, y_val_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        avg_accuracy = 100 - (np.abs(y_val - y_val_pred) / y_val).mean() * 100
        min_accuracy = 100 - (np.abs(y_val - y_val_pred) / y_val).max() * 100

        print(f"\nModel: {model_name} (Validation Results)")
        print(f"Validation MSE: {val_mse}")
        print(f"Validation MAE: {val_mae}")
        print(f"Validation R²: {val_r2}")
        print(f"Average Accuracy: {avg_accuracy:.2f}%")
        print(f"Minimum Accuracy: {min_accuracy:.2f}%")

        # Retrain on training + validation set and test on test set
        X_train_val = np.vstack((X_train, X_val))
        y_train_val = np.concatenate((y_train, y_val))
        model.fit(X_train_val, y_train_val)
        y_test_pred = model.predict(X_test)

        # Test metrics
        test_mse = mean_squared_error(y_test, y_test_pred)
        test_mae = mean_absolute_error(y_test, y_test_pred)
        test_r2 = r2_score(y_test, y_test_pred)
        test_avg_accuracy = 100 - (np.abs(y_test - y_test_pred) / y_test).mean() * 100
        test_min_accuracy = 100 - (np.abs(y_test - y_test_pred) / y_test).max() * 100

        print(f"\nModel: {model_name} (Test Results)")
        print(f"Test MSE: {test_mse}")
        print(f"Test MAE: {test_mae}")
        print(f"Test R²: {test_r2}")
        print(f"Average Accuracy: {test_avg_accuracy:.2f}%")
        print(f"Minimum Accuracy: {test_min_accuracy:.2f}%")

          

                # Plot the actual vs predicted values for the test set
        plt.figure(figsize=(10, 5))

        # Use evenly spaced indices for plotting
        indices = range(len(y_test))
        plt.plot(indices, y_test.reset_index(drop=True), label="Actual", marker="o")
        plt.plot(indices, y_test_pred, label="Predicted", linestyle="--", marker="x")

        # Use 'test_dates' as x-tick labels
        plt.xticks(indices, test_dates, rotation=45)  # Map indices to test_dates

        # Add labels, title, legend, and grid
        plt.title(f"Actual vs Predicted for {model_name} on Test Set")
        plt.xlabel("Test Dates")
        plt.ylabel("Target Value")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()  # Ensure layout fits without overlap

        plt.show()





    dataset_results[model_name] = f"{avg_accuracy:.2f}% / {min_accuracy:.2f}%"
    if model_name not in best_models or avg_accuracy > best_models[target_column]["accuracy"]:
                best_models[target_column] = {
                    "model": model,
                    "scaler": scaler,
                    "poly": poly,
                    "log_transform": log_transform,
                    "accuracy": avg_accuracy,
                    "name": model_name,
                }
        
    summary_results.append(dataset_results)

# Save summary results
summary_df = pd.DataFrame(summary_results)
summary_df.to_csv("summary_results.csv", index=False)

# Test the best models on new test sets
for dataset_path, target_column, test_path in zip(dataset_paths, target_columns, test_paths):
    # Load new test set
    test_data = pd.read_csv(test_path)
    test_data.fillna(test_data.mean(), inplace=True)

    # Drop unnecessary columns and align with training data
    X_test = test_data.drop(columns=[target_column], errors='ignore')  
    y_test = test_data[target_column]

    # Retrieve the best model information
    best_model_info = best_models[target_column]
    model = best_model_info["model"]
    scaler = best_model_info["scaler"]
    poly = best_model_info["poly"]
    log_transform = best_model_info["log_transform"]
    model_name = best_model_info["name"]

    # Apply log-transform if it was used during training
    if log_transform:
        y_test = np.log1p(y_test)

    # Apply the same preprocessing to the test data
    # Drop any columns not present in training (ensure alignment)
    X_test = X_test[scaler.feature_names_in_]  # Select only features used during training

    # Scale and transform the test data
    X_test_scaled = scaler.transform(X_test)
    X_test_poly = poly.transform(X_test_scaled)

    # Predict on test data
    y_test_pred = model.predict(X_test_poly)

    # Reverse log-transform if it was applied
    if log_transform:
        y_test_pred = np.expm1(y_test_pred)
        y_test = np.expm1(y_test)

    # Calculate metrics
    avg_accuracy = 100 - (np.abs(y_test - y_test_pred) / y_test).mean() * 100
    min_accuracy = 100 - (np.abs(y_test - y_test_pred) / y_test).max() * 100
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    # Save results to a CSV
    results = pd.DataFrame({
        "Model": model_name,
        "Actual Target": y_test.values,
        "Predicted Target": y_test_pred,
        "Accuracy (%)": 100 - (np.abs(y_test - y_test_pred) / y_test * 100),
    })
    results.to_csv(f"test_results_{target_column}.csv", index=False)

    # Print results
    print(f"\nTest Results for {target_column}:")
    print(f"Model: {model_name}")
    print(f"Test MSE: {test_mse}")
    print(f"Test R²: {test_r2}")
    print(f"Average Accuracy: {avg_accuracy:.2f}%")
    print(f"Minimum Accuracy: {min_accuracy:.2f}%")
    print(f"Results saved to test_results_{target_column}.csv")

print("Processing complete. Summary and test results saved.")
```
